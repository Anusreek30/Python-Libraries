```python
#Basic Line plot
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title(' Basic Line Plot')

plt.plot(x,y)
plt.show
```

```python
#Line graph with Markers
import matplotlib.pyplot as plt
import numpy as np
x=[0,1,2,3,4]
y=[0,1,4,9,16]
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot with Markers')
plt.plot(x,y,marker='o')
plt.show
```

TASK 3

```python
#Multiple Line PLots
import matplotlib.pyplot as plt
import numpy as np
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Multiple Line Plots')
plt.plot(x1,y1,color='g',linestyle='dashed')
x2= [1, 2, 3, 4]
y2 = [2, 3, 6, 8]
plt.plot(x2,y2,color='r',linestyle='solid')
plt.legend(['green','red'],loc='lower right')
plt.show()
```

TASK 4

```python
#Subplots with Line Graphs
import matplotlib.pyplot as plt
import numpy as np
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Subplots with Line Graphs')
plt.subplot(1, 2, 1)
plt.plot(x, y, color='g', linestyle='dashed')
plt.show()
plt.figure()
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.subplot(1, 2, 2)
plt.plot(x, y, color='r', linestyle='solid')
plt.show()
```

TASK 7

```python
#Bar Plot with Annotations
import matplotlib.pyplot as plt
import numpy as np
x = np.array(['x','y','z'])
y = np.array([7,2,5])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot with Annotations')
plt.bar(x, y, color='g')
plt.annotate('7', (x[0], y[0]), textcoords="offset points", xytext=(0, 10))
plt.annotate('2', (x[1], y[1]), textcoords="offset points", xytext=(0, 10))
plt.annotate('5', (x[2], y[2]), textcoords="offset points", xytext=(0, 10))
plt.show()
```

```python
#Create Two Figures with Different Plots
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4]
y=[1,4,9,16]
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot')
plt.plot(x,y,color='r')
plt.show()
plt.figure()
labels = ['A', 'B', 'C']
values = [5, 7, 9]
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.bar(labels, values, color='g')
plt.show()
```

```python
#Adjust the Figure Size
import matplotlib.pyplot as plt
import numpy as np
x = [1, 2, 3, 4, 5]
y = [5, 3, 8, 7, 4]
plt.figure(figsize=(10, 6))


plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot')
plt.plot(x, y, color='r')

plt.show()
```
